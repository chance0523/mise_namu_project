```python
import pandas as pd
from collections import OrderedDict
```

### 원본 파일에서 필요한 데이터만 가져와 df0 만들기

```python
df0 = pd.read_csv("자치구단위 서울 생활인구 일별 집계표2.csv")
df0 = df0.loc[ 0:7045, ["기준일ID", "시군구코드", "시군구명", "총생활인구수"]]
df0
```

```python
# 시군구코드 리스트
code_list = df0["시군구코드"]
code_list = list(OrderedDict.fromkeys(code_list))

# 시군구명 리스트
region_list = df0["시군구명"]
region_list = list(OrderedDict.fromkeys(region_list))

# 시군구코드에 해당하는 시군구명 딕셔너리 만들기
dic = {}

for a in range(0, len(code_list)):
    dic[code_list[a]] = region_list[a]
    
dic
```

### 결과 값을 저장할 데이터프레임 df1 만들기

```python
# df1 컬럼에 사용할 month_list
month_list = []
month_list.append('시군구명')

for month in range(4, 13):
    month_list.append(f'{month}월')

# df1 컬럼에 연 합계, 연 평균 추가
month_list.append("연 합계")
month_list.append("연 평균")
```

```python
df1 = pd.DataFrame(columns = month_list, index = code_list)
df1
```

### 원본 데이터에서 월별 인구수 계산하기

```python
for code in code_list:
    df2 = df0.loc[ df0["시군구코드"]==code, ["기준일ID","시군구코드", "총생활인구수"]]
    
    # df2의 인덱스를 재설정 (0부터 시작하도록!)
    df2 = df2.reset_index(drop=True)
    
    # df2의 기준일을 날짜 형식으로 변환
    df2["기준일ID"] = pd.to_datetime(df2["기준일ID"].astype(str), format='%Y-%m-%d')
    
    total = 0
    year_total = 0
    
    # 월별 마지막 인덱스가 다름
    # 기준일ID의 월에 따라 df1에 합계 저장
    for i in range(0,df2.index[-1]):
        if df2["기준일ID"][i].month == 4:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "4월"] = total
        
        elif df2["기준일ID"][i].month == 5:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "5월"] = total

        elif df2["기준일ID"][i].month == 6:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "6월"] = total
    
        elif df2["기준일ID"][i].month == 7:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "7월"] = total
    
        elif df2["기준일ID"][i].month == 8:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "8월"] = total
    
        elif df2["기준일ID"][i].month == 9:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "9월"] = total
    
        elif df2["기준일ID"][i].month == 10:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "10월"] = total
    
        elif df2["기준일ID"][i].month == 11:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "11월"] = total
    
        elif df2["기준일ID"][i].month == 12:
            total = total + df2["총생활인구수"][i]   
            df1.loc[code, "12월"] = total
    
    # 시군구별 연 합계와 연 평균을 구함
    for j in range(4, 13):
        year_total = year_total + df1.loc[code, f'{j}월']
        
    df1.loc[code, "시군구명"] = dic[code]
    df1.loc[code, "연 합계"] = year_total
    df1.loc[code, "연 평균"] = year_total / 9
    
df1
```

### 데이터 파일로 저장

```python
df1.to_csv("living_population.csv", encoding="utf-8-sig")
```
