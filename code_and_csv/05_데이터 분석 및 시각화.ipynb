```python
import pandas as pd
```

### 미세먼지 데이터

```python
df_dust_1 = pd.read_csv('mise_namu_project/airkorea/미세먼지 평균 데이터.csv', encoding='utf-8-sig')
df_dust_1 = df_dust_1.loc[ :, ["시군구명", "4월~12월"]]

# 시군구명 오름차순 정렬
df_dust_1.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df_dust_1
```

```python
# 구별로 데이터 합치기
df_dust_group = df_dust_1["4월~12월"].groupby(df_dust_1["시군구명"])
```

```python
df_dust = pd.DataFrame(df_dust_group.mean())
df_dust
```

### 생활인구 데이터

```python
# living_population (생활인구) 파일 불러오기
df_population = pd.read_csv('living_population.csv', encoding='utf-8-sig')
df_population = df_population.loc[ 1: ,["시군구명", "연 평균"]]
df_population.sort_values(by=["시군구명"], ascending=['True'], inplace=True)
df_population
```

### 교통량 데이터

```python
df_traffic_1 = pd.read_csv('서울시_교통량_경기도제외.csv', encoding='utf-8-sig')
df_traffic_1 = df_traffic_1.loc[ :, ["시군구명", "합계"]]
```

```python
# 시군구명 오름차순 정렬
df_traffic_1.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df_traffic_1
```

```python
# 구별로 데이터 합치기
df_traffic_group = df_traffic_1["합계"].groupby(df_traffic_1["시군구명"])
```

```python
df_traffic = pd.DataFrame(df_traffic_group.sum())
df_traffic
```

### 난방

```python
df_gas = pd.read_csv('서울시_구별_난방.csv', encoding='utf-8-sig')
df_gas = df_gas.loc[ : ]

# 시군구명 오름차순 정렬
df_gas.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df_gas
```

### 전력

```python
df_elec = pd.read_csv('서울시_구별_전력.csv', encoding='utf-8-sig')
df_elec = df_elec.loc[ : ]

# 시군구명 오름차순 정렬
df_elec.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df_elec
```

### 경유차 데이터

```python
df_car = pd.read_csv('구별_차량등록_현황.csv', encoding='utf-8-sig')
df_car = df_car.loc[ : ]

# 시군구명 오름차순 정렬
df_car.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df_car
```

### 공사현황

```python
df_con = pd.read_csv('서울시_공사현황.csv', encoding='utf-8-sig')
df_con = df_con.loc[ : ]

# 시군구명 오름차순 정렬
df_con.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df_con
```

### 도로굴착

```python
df_road = pd.read_csv('서울시_도로굴착.csv', encoding='utf-8-sig')
df_road = df_road.loc[ : ]

# 시군구명 오름차순 정렬
df_road.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df_road
```

### 데이터셋 합쳐서 데이터프레임 만들기

```python
# 구 이름 리스트로 만들기
name_list = list(df_road["시군구명"])
name_list
```

```python
# 데이터프레임 생성
df = pd.DataFrame(index = name_list)
```

```python
# 초기 데이터프레임 merge
df = pd.merge(df_dust, df_population, on="시군구명", how="inner")
df = pd.merge(df, df_traffic, on="시군구명", how="inner")
df = pd.merge(df, df_gas, on="시군구명", how="inner")
df = pd.merge(df, df_elec, on="시군구명", how="inner")
df = pd.merge(df, df_car, on="시군구명", how="inner")
df = pd.merge(df, df_con, on="시군구명", how="inner")

df.columns = ["시군구명", "미세먼지", "생활인구", "교통량", "난방", "전력", "경유차", "건설현장"]
```

```python
df.corr()
```

```python
# 상관관계 파악 후 필요한 요인만 넣은 데이터프레임 merge
df = pd.merge(df_dust, df_traffic, on="시군구명", how="inner")
df = pd.merge(df, df_gas, on="시군구명", how="inner")
df = pd.merge(df, df_car, on="시군구명", how="inner")

df.columns = ["시군구명", "미세먼지", "교통량", "난방", "경유차"]
```

```python
# df의 인덱스를 시군구명 컬럼으로 바꾸기
# 데이터가 숫자로만 이루어져있어야 히트맵 가능
column_list = ["미세먼지", "교통량", "난방", "경유차"]
```

```python
df_number = df.set_index("시군구명")
df_number
```

### 정규화

```python
from sklearn import preprocessing
col = ['미세먼지','교통량','난방','경유차']
x = df[col].values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x.astype(float))
df_norm = pd.DataFrame(x_scaled, columns=col, index = df['시군구명'])
import numpy as np
df_norm['평균'] = np.mean(df_norm[col[1:]],axis=1)
df_norm
```

### 시각화 (히트맵 그리기)

```python
plt.figure(figsize=(10, 10))
plt.title("상관관계 분석", fontsize=20)

ax = sns.heatmap(df_norm, annot=True, cmap = 'YlGnBu')

bottom, top = ax.get_ylim()
ax.set_ylim(bottom+0.5, top-0.5)

plt.show()
```
