```python
import pandas as pd
```

### 생활인구 데이터

```python
# living_population (생활인구) 파일 불러오기
df1 = pd.read_csv('living_population.csv', encoding='utf-8-sig')
df1 = df1.loc[ 1: ,["시군구명", "연 평균"]]
df1.sort_values(by=["시군구명"], ascending=['True'], inplace=True)
df1
```

```python
# 구 이름 리스트로 만들기
name_list = list(df1["시군구명"])
name_list
```

### 교통량 데이터

```python
# 교통량 데이터 가져오기
df2 = pd.read_csv('monthly_traffic_sum.csv', encoding='utf-8-sig')
df2 = df2.loc[ :, ["시군구명", "연 평균"]]

# 시군구명 오름차순 정렬
df2.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df2
```

```python
# 구별로 데이터 합치기
df2_group = df2["연 평균"].groupby(df2["시군구명"])
```

```python
df2_mean = pd.DataFrame(df2_group.mean())
df2_mean
```

### 미세먼지 데이터

```python
df3 = pd.read_csv('mise_namu_project/airkorea/미세먼지 평균 데이터.csv', encoding='utf-8-sig')
df3 = df3.loc[ :, ["시군구명", "연 평균"]]

# 시군구명 오름차순 정렬
df3.sort_values(by=["시군구명"], ascending=[True], inplace=True)
df3
```

```python
# 구별로 데이터 합치기
df3_group = df3["연 평균"].groupby(df3["시군구명"])
```

```python
df3_mean = pd.DataFrame(df3_group.mean())
df3_mean
```

### 데이터셋 3개 한 데이터프레임에 저장하기

```python
# 데이터프레임 생성
df = pd.DataFrame(index = name_list)
```

```python
df = pd.merge(df3_mean, df2_mean, on="시군구명", how="inner")
df_final = pd.merge(df, df1, on="시군구명", how="inner")
df_final.columns = ["시군구명", "미세먼지", "교통량", "생활인구"]
df_final
```
