```python
# 데이터프레임 인덱스, 파일 불러오기에 사용할 station list
hangul_station_list=[]
for station_list in hangul_station_Dict.keys():
    hangul_station_list.append(station_list)
```

```python
# 데이터프레임 컬럼에 사용할 month list
month_list = []
for month in range(1, 13):
    month_list.append(f'{month}월')
    
# 데이터프레임 컬럼에 연 평균 추가
month_list.append("연 평균")
```

```python
# 인덱스가 지역, 컬럼명이 월인 빈 데이터프레임 생성
df = pd.DataFrame(columns = month_list, index = hangul_station_list)
df
```

```python
# 관측소별 미세먼지 농도 월평균 구해서
# 데이터프레임에 저장하기

for name in hangul_station_list:
    mean_sum = 0
    year_avg = 0
    
    for month in range(1,13):      
        df1 = pd.read_csv(f'station_list/{name}/2018/{name}_2018_{month}.csv')
        mean = df1["pm10"].mean()
        mean_sum = mean_sum + mean
        df.loc[name, f'{month}월'] = mean
        
    year_avg = mean_sum / 12
    df.loc[name, "연 평균"] = year_avg
```
